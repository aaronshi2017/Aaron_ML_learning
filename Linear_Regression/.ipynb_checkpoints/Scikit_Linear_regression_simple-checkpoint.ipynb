```python
# This is trying to study how to use sklearn_linearRegression method to do linear regression prediction, to evaluate the time spent and the mean square error. 


import numpy as np
import time
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
%matplotlib inline

X=np.random.uniform(1,10,1000) # construct 1000 samples between (1,10)
# print(X)
W_=np.random.normal(loc=4,scale=0.5,size=1000) # using normalized distribution values to represent the slope W_
b_=np.random.normal(loc=5,scale=3,size=1000) # using normlized distribution values to represent the intercept b_
Y=np.multiply(W_,X)+b_  #construct Y as trained data 
start=time.time()
reg=LinearRegression().fit(X.reshape(-1,1),Y) # call LinearRegression to fit()
end=time.time()

W=reg.coef_ # the slope by calculation
b=reg.intercept_ # the intercept by calculation
spenttime=end-start

def linear_regression(X,w,b):
    return w*X+b

def mean_square(y_pred,y_value):
    return mean_squared_error(y_pred,y_value)

# use diagram to show the performance of Sklearn-method
plt.scatter(X, Y, c='r', label="original data")
pred=linear_regression(X,W,b)
plt.plot(X, pred, c='g', label='Fitted line')
loss=mean_square(pred,Y)
plt.title("Used Time: %.2f"%spenttime+'---W: %.2f'%W+'---b: %.2f'%b+'---Loss: %.2f'%loss)
plt.show()

```
